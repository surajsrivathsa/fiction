```python
gfp = "/Users/surajshashidhar/git/german-emotion-dictionary/fundamental"
df_disgust = pd.read_csv(gfp+"/Ekel.txt",header=None)
df_disgust["emotion"] = "disgust"

df_joy = pd.read_csv(gfp+"/Freude.txt",header=None)
df_joy["emotion"] = "joy"

df_fear = pd.read_csv(gfp+"/Furcht.txt",header=None)
df_fear["emotion"] = "fear"

df_sadness = pd.read_csv(gfp+"/Trauer.txt",header=None)
df_sadness["emotion"] = "sadness"

df_surprise = pd.read_csv(gfp+"/Ueberraschung.txt",header=None)
df_surprise["emotion"] = "surprise"

df_contempt = pd.read_csv(gfp+"/Verachtung.txt",header=None)
df_contempt["emotion"] = "contempt"

df_anger = pd.read_csv(gfp+"/Wut.txt",header=None)
df_anger["emotion"] = "anger"
```
