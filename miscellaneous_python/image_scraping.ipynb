```python
#!/usr/bin/python3
#-*- coding:utf-8 -*-

import re
import requests
from bs4 import BeautifulSoup
from urllib.parse import urlparse
import os
import pandas as pd
```

```python
def download_google(pgid, book_name, auth_name, dest_path):
    query_str = book_name + " " + auth_name + " book cover"
    url = 'https://www.google.com/search?q=' + query_str + '&client=opera&hs=cTQ&source=lnms&tbm=isch&sa=X&ved=0ahUKEwig3LOx4PzKAhWGFywKHZyZAAgQ_AUIBygB&biw=1920&bih=982'
    print(url)
    page = requests.get(url).text
    soup = BeautifulSoup(page, 'html.parser')
    
    lst = soup.find_all("img", limit = 5)
    if lst:
        flag = False
        for raw_img in lst:
            link = raw_img.get('src')
            #print(link)
            if(link.startswith("https") and flag == False):
                os.system('wget -c ' + "'" + link + "'" + ' -O '+  dest_path + pgid + ".jpg")
                flag = True
                break;
        if(flag == False):
            print("{} , {} Image was undownloadable as https link was unavailable, download it manually".format(pgid, book_name))
            
    else:
        print("{} , {} Image not found on internet".format(pgid, book_name))
    
    """
    for raw_img in soup.find_all('img'):
        link = raw_img.get('src')
        os.system("wget -c " + link + " -O "+ "/Users/surajshashidhar/Desktop/ovgu/semester_2/XAI_project/reasearched_code_and_data/book_covers/" + "pride_and_prejuidice.jpg")
    """
```
